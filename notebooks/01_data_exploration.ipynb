```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ticker_symbol = "GC=F"
```

```python
print(f"Downloading {ticker_symbol}...")
gold_data = yf.download(ticker_symbol, start="2015-01-01", end="2025-12-31")
```

```python
print(gold_data.head())
```

```python
plt.figure(figsize=(12, 6))
plt.plot(gold_data['Close'], label="Gold Price (Close)")
plt.title("Gold price chart from 2015 - Present")
plt.xlabel("Year")
plt.ylabel("Price (USD)")
plt.legend()
plt.show()
```

```python
gold_data.to_csv(r'..\data\raw\gold_price_dataset.csv')
```
